#### Ejercicio 4
Escribir el código en Python que solicite el correo de todos los empleados los muestre por pantalla

```python
import sqlite3 as sql_bd # Se importara la libreria sqlite3

'''
Realizaremos la conexion a la base de datos SQLite para 
ello se usara la funcion connect(ruta) para conectarse a 
la base de datos e interactuar con ella a traves del objeto
de tipo connect que devuelve la funcion.

Tienes que agregar la ruta relativa desde la carpeta raiz
del proyecto de Python, en este caso la carpeta raiz se
llama "Python" que tenia dos subcarpetas Tarea1, Tarea2...
'''
conexion = sql_bd.connect('Chinook_Sqlite.sqlite')

'''
Crearemos un objeto de tipo cursor con la funcion cursor() y 
se configurara dicho objeto para que realize la consulta a 
la base de datos con cursor.execute(consulta)
'''
cursor = conexion.cursor()
cursor.execute('select Email from Employee')

'''
Con fetchall() se pedira los resultados de la consulta y 
se almacenaran en una variable
'''
resultado = cursor.fetchall()

'''
Se imprimira el resultado de tal manera que todas las filas tengan 
salto de linea. Puesto que resultado se ofrece esta en formato de
lista y esta no tiene saltos de linea.
'''
print(*resultado, sep='\n') 
'''
Con la funcion close() se cerrara la conexion, cuando no se necesita 
la base de datos hay que cerrarla 
'''
conexion.close()
```

#### Ejercicio 5

#### Ejercicio 7
Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie (pd.Series) con las notas de los alumnos aprobados ordenadas de mayor a menor.

```python
import pandas as pd # Primero importamos la libreria pandas

# Segundo se creara el diccionario con los datos
notas_curso = {"Juan Manuel Castilla Fernandez": 9,
         "Alonso Diaz Martinez": 8,
         "Lucia Palomares": 7,
         "Daniel Serrano": 3,
         "Andrea Reviejo": 2,
         "Sofia Castilla Sanchez": 10,
         "Ismael Alba Castellon": 7,
         "Nerea Rosales": 6
         }

# Tercero se declara la funcion con su parametro
def ordenar_notas(notas_recopiladas):
    ''' 
    Quinto se programa la funcion para que empiece 
    a ordenar los valores del diccionario
    '''
    ordenacion = pd.Series(notas_recopiladas) # Se crea la Serie con el diccionario pasado por parametro
    return ordenacion[ordenacion >= 5].sort_values(ascending=False) # y se retornara todos los valores superiores a 5 y con short_values se ordenaran los valores 


notas_ordenadas = ordenar_notas(notas_curso) # Cuarto se llama a la funcion y se guardara el resultado en una variable
print(notas_ordenadas)
```

#### Ejercicio 8
Escribir una función que construya un DataFrame a partir del fichero cotizacion.csv y devuelva otro DataFrame con el mínimo, el máximo y la media de cada columna.

```python
import pandas as pd

def analisis_cotizacion(csv):
    '''
    Retornara un objeto DataFrame, 
    del archivo csv que se pase 
    por parametro
    '''
    return pd.DataFrame(
            pd.read_csv(csv, sep=';'), # Leeara el fichero csv, teniendo en cuenta que el delimitador de este fichero es ;
            columns=['Final', 'Máximo', 'Mínimo'] # Una vex leido el csv, se proyectara las columnas 'Final', 'Máximo' y 'Mínimo'
        )

print(analisis_cotizacion('cotizacion.csv'))
```

#### Ejercicio 9
Crear un df desde el archivo automobile_data.csv y encontrar la media de precio por fabricante

```python
import pandas as pd

datos_importados_automoviles = pd.read_csv('Automobile_data.csv', sep=',') # Se leera el archivo csv, teniendo en cuenta que su delimitador es ,
automoviles = pd.DataFrame(datos_importados_automoviles) # Se creara el objeto DataFrame con el csv leido

# Se pedira una lista de datos de la columna company con loc, eliminando los datos duplicados duplicados y se meteran dentro de una lista
list_companies = list(automoviles.loc[:, 'company'].drop_duplicates())

# Se creara un bucle for para iterar la lista anterior y se ejecutara lo siguiente
for company_now in list_companies:
    '''
    Se hara una consulta al DataFrame creado anteriormente
        con el metodo query se creara una consulta tipo WHERE para filtrar lineas
        donde en este caso se pide todas las lineas de la tabla donde las celdas 
        de la columna "company" tengan el valor actual de la variable company_now
    '''
    precio = automoviles.query(('company == \'' + company_now + '\'')).loc[:, 'price'].mean()
    print('La media de la compañia', company_now, 'tiene una media de', precio, 'euros.') # Se imprimira el resultado
```

#### Ejercicio 10
A partir del archivo annual.csv crear un DataFrame. Representar gráficamente en una figura de líneas los datos del df

```python
import pandas as pd # Se llamaran las libreria pandas y matplotlib
import matplotlib.pyplot as mp

anual = pd.read_csv('annual.csv', sep=',') # Se leera el csv teniendo encuenta que el delimitador de este archivo es ,
anual_DataFrame = pd.DataFrame(anual) # Se creara un DataFrame a partir del archivo csv leido

'''
Se pedira los valores de la columan Source, 
pero eliminando duplicados, y se agregara a 
una lista
'''
lista_fuentes = list(anual.loc[:, 'Source'].drop_duplicates())

region, axes = mp.subplots() # Se usara la funcion subplot para crear una ventana para graficas 

# Se creara un bucle for iterando la lista anterior para iterar con la lista
for fuente in lista_fuentes:
    '''
    Se hara un filtrado de todas las lineas donde las celdas de la columna Source
    tenga el mismo valor que la variable del bucle
    '''
    consulta = anual_DataFrame.query(('Source == \'' + fuente + '\''))
    consulta.plot(x="Year", y="Mean", ax=axes) # Se creara la grafica con los datos optenidos de la consulta

axes.legend(lista_fuentes) # Se agregara una leyenda
mp.show() # Se visualizara la grafica
```

#### Ejercicio 11
Con el df anterior, representar en una figura dos gráficos de líneas dependiendo de la fuente (GCAG o GISTEMP) con colores diferentes, leyenda, títulos individuales y un título global

```python
import pandas as pd
import matplotlib.pyplot as mp

# Se creara las variables que se necesitan
cont = 0
lista_colores = ['blue', 'orange']

# Se creara el Dataframe a partir de los datos obtenidos del csv
anual = pd.read_csv('annual.csv', sep=',')
diccionario_DataFrame = pd.DataFrame(anual)

# Se creara una lista con los datos de la columna Source
lista_fuentes = list(anual.loc[:, 'Source'].drop_duplicates())

# Se creara una ventana para diseñar graficas necesarios, en este caso seran dos graficas
region, axes = mp.subplots(1,2, sharey=True, layout='constrained')
# Se fijara el ancho de la grafica
region.set_figwidth(12)

# Se creara un bucle for que itere la lista de las fuentes
for fuente in lista_fuentes:
    '''
    Se hara un filtrado de todas las lineas donde las celdas de la columna Source
    tenga el mismo valor que la variable del bucle
    '''
    consulta = diccionario_DataFrame.query(('Source == \'' + fuente + '\''))
    # Se proyectara la columna Year y Mean y se guaradara la informacion de cada columna en dos variables
    year = list(consulta.loc[:, 'Year'])
    mean = list(consulta.loc[:, 'Mean'])
    ''' 
    Se creara un grafico de linea para cada fuente, con los datos proporcionados, 
    teniendo en cuenta que tienen que ser de colores diferentes
    '''
    axes[cont].plot(year, mean, color='tab:'+lista_colores[cont])
    ''' 
    Se creara una leyenda para cada grafico de linea que estaran arriba a la izquierda

    NOTA: es importante declarar el contenido de la variable del for como si fuera una
    lista y no una cadena, sino mostrara solamente el primer caracter del texto pasado
    por el primer parametro, y la leyenda solamente tendra la primera letra del texto,
    y no el texto completo.
    '''
    axes[cont].legend([fuente], loc='upper left')
    # Se agregara un titulo a cada grafica
    axes[cont].set_title('Bloque de ventas ' + fuente)
    '''
    Se creara un contador para poder interactuar con el array que permite interactur 
    con los cuadros de cada grafica
    '''
    cont += 1 

# Se agregara un titulo global
region.suptitle('Bloques de ventas historicas', fontsize=28)

mp.show()
```

#### Ejercicio 13
Utilizando la librería tkinter y etiquetas de texto dibujar el resultado final de una partida de 3 en raya (X y O) en una ventana con un grid de 3x3.

```python
from tkinter import *

mi_ventana = Tk() # Se instanciara el objeto Tk()
# Se crearan las variables que se necesiten
textos = list()
texto_actual = ''
cont_column = 0
cont_row = 0

# Se declarara el titulo y las dimensiones de la ventana
mi_ventana.title("El juego de las Tres en Raya")
mi_ventana.geometry('720x480')

''' 
Se creara un bucle for para imprimir las fichas de los jugadores

Para ello dentro del bucle se creado una condicional para que elija
el texto que se le proporcionara a cada Label, y se crearan 9 labels
y se introducciran dentro de una lista
'''
for cont in range(1,10):
    if cont in [1,8,9]:
        texto_actual = ''
    elif cont % 2 == 0:
        texto_actual = 'X'
    else:
        texto_actual = 'O'
        
    textos.append(Label(mi_ventana, text=texto_actual, font={"Arial", 20}))

# Se creara un grid que contrendran los Label creados anteriormente
for main_cont in range(0,9):
    textos[main_cont].grid(column=cont_column, row=cont_row)
    
    if cont_column == 2:
        cont_column = 0
        cont_row += 1
    else:
        cont_column += 1

# Se creara el Label que indica al ganador de la partida
texto_ganador = Label(mi_ventana, text="El ganador a sido el jugador con las fichas O", font={"Arial", 20})
texto_ganador.grid(column=5, row=0, columnspan=3)

mainloop() # Se ejecutara la ventana Tk()
```

#### Ejercicio 14
Utilizando la librería *random* de python, crear una ventana con un botón que al pulsarlo devuelva un número aleatorio del 1 al 20

```python
from tkinter import *
import random

# Se instanciara el objeto Tk()
mi_ventana = Tk()

# Se declarara el titulo y las dimensiones de la ventana
mi_ventana.title("Numero aleatorio")
mi_ventana.geometry('720x480')

''' 
Se creara un label en el cual se generara un numero aletorio, zfill es para
que en caso de que el numero generado sea de un digito tenga un cero a la izquierda
'''
numero = str(random.randrange(1,20)).zfill(2)
texto = Label(mi_ventana, text=numero, font={"Arial", 20})
texto.grid(column=0,row=0) # Introccimos el Label

# Se creara una funcion para generar un nuevo label con un nuevo numero aleatorio
def numero_aleatorio():
    numero = str(random.randrange(1,20)).zfill(2)
    texto = Label(mi_ventana, text=numero, font={"Arial", 20})
    texto.grid(column=0,row=0)

# Se agregara un boton que use la funcion anterior
boton = Button(mi_ventana, text = "Generar numero", command = numero_aleatorio)
boton.grid(column=0, row=1)

mainloop()
```
