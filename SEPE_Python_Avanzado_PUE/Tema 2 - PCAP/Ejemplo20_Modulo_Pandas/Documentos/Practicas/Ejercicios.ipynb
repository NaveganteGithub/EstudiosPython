## Procesar datos

```python
# Me quedo con las url que no son nulas
serieUrlLanding = gclid_10["url_landing"][gclid_10["url_landing"].notnull()]
# con la url, utilizamos split=& y con cada fragmento genero una nueva columna
gclid_10[['url_base','idUser','uuid_url','camp','adg','device','sl','adv','rec','vacia' ]] = (
    serieUrlLanding.str.split(pat="&",n=10, expand=True))
```

```python
# Eliminaremos las columnas que no necesitaremos
gclid_colum = gclid_10.drop(['idUser', 'uuid_url', 'rec', 'vacia'], axis=1)
gclid_colum
```

```python
# Cuando quieras aplicar instrucciones python a los dataframes utiliza la funcion apply es mucho mas manejable 
url = gclid_colum['url_base'].apply(lambda url: str(url).split("?")[0])
# url = url.apply(lambda url: str(url)[str(url).rfind("/") + 1:])

def obtenerModelo(url):
    if str(url) == 'nan': 
        return 'home'
    mi_url = str(url).split("?")[0]
    posicion = mi_url.rfind('/')
    modelo = mi_url[posicion + 1:]
    if modelo == '' or modelo == 'renting':
        modelo = 'home'
    return modelo
    
gclid_colum.insert(loc=7, column='modelo', value= gclid_colum['url_base'].apply(obtenerModelo))    
print(gclid_colum['modelo'].head(50))
```

## Crear un orden de usuarios
<p>Como tenemos usuarios repetidos tendria un problema al hacer merge (union) con conversiones.csv, por ello vamos a eliminar con los usuarios que no estan repetidos</p>

```python
# Crear una columna llamada orden que almacene el orden de los id_user
id_user_order = gclid_colum.sort_values(['id_user','ts'])
id_user_order.insert(loc=1, column='orden', value=id_user_order.groupby('id_user')['ts'].cumcount())
```

```python
id_user_order = id_user_order[id_user_order['orden'] == 0]
# id_user_order['orden'].head(50)
id_user_order['orden'].shape
```

## Conversiones

```python
conversiones = pd.read_csv("../Recursos/conversiones.csv", sep=";")
conversiones
```

```python
df_final = pd.merge(id_user_order, conversiones, how='outer', on=['id_user'])
df_final
```

```python
df_final.insert(loc=1, column='conversiones', value=df_final['lead_type'].apply(lambda conversion: 1 if str(conversion) in ['CALL', 'FORM'] else 0))
df_final['conversiones'].value_counts().head(50)
```

```python
df_columnas_elim = df_final.drop(['orden', 'url_landing', 'url_base'], axis=1)
df_columnas_elim.columns
```
